# Skewness and Outlier removal:

```python
data.skew()
```

Keeping +/- 0.5 as the range for skewness.

We have some columns which are going beyond the above range.

```python
#removing outliers
from scipy.stats import zscore
z=np.abs(zscore(data))
z
```

```python
#threshold=3
print(np.where(z>3))
```

```python
data_new=data[(z<3).all(axis=1)] #removing outliers
```

```python
data.shape
```

```python
data_new.shape
```

```python
#Percentage data loss:
loss_percent=(30-29)/30*100
print(loss_percent)
```

After removing outliers, we have lost 03.33% data.

```python
x=data_new.drop('Wins', axis=1)
y=data_new['Wins']
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import power_transform
data_new=power_transform(x)

data_new=pd.DataFrame(data_new,columns=x.columns)
```

```python
data_new.skew()
```

We can see, now skewness has been reduced.

# Now we will try to find out the best random state

```python
maxAcc=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    Ln=LinearRegression()
    Ln.fit(x_train,y_train)
    pred=Ln.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
print("Best accuracy is ",maxAcc, " at Random State ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=maxRS)
```

```python
model=[LinearRegression(),Lasso(alpha=0.0001),Ridge(alpha=0.0001),ElasticNet(alpha=0.0001)]
for m in model:
    m.fit(x_train,y_train)
    #sc=m.score(x_train,y_train)
    predm=m.predict(x_test)
    acc=r2_score(y_test,predm)
    print('Accuracy Score of',m,'is:',acc)
    print('mean_absolute_error:',mean_absolute_error(y_test,predm))
    print('mean_squared_error:',mean_squared_error(y_test,predm))
    print('Root mean_squared_error:',np.sqrt(mean_squared_error(y_test,predm)))
    print("\n")
```

# Cross Validation of model:

```python
model=[LinearRegression(),Lasso(alpha=0.0001),Ridge(alpha=0.0001),ElasticNet(alpha=0.0001)]
for m in model:
    score=cross_val_score(m,x,y,cv=5)
    print("Score for",m,"is: ",score.mean())
```

With ElasticNet model there is very less difference in accuracy and cross validation score, Hence the best model is ElasticNet.

# Hyper Parameter Tunning:

```python
from sklearn.model_selection import GridSearchCV
```

```python
#creating  parameter list to pass in GridSearchCV
#ElasticNet
param={'alpha':[1.0,0.1,0.001,0.0001],'fit_intercept':[True,False], 'normalize':[True,False],
       'copy_X':[True, False],'precompute':[True,False],'max_iter':[250,500,1000],
       'tol':[1.0,0.1,0.001,0.0001],'warm_start':[True,False],'positive':[True,False],
       'random_state':range(1,20)}
```

```python
#GCV=GridSearchCV(Lasso(),param,cv=5)
GCV=GridSearchCV(ElasticNet(),param,cv=5)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
GCV_pred=GCV.best_estimator_.predict(x_test) 
GCV_pred
```

```python
r2_score(y_test,GCV_pred)
```

After performing GridSearchCV method accuarcy is 81.28%
