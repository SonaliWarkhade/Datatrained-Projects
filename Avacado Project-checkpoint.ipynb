# Encoding:

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()

#data['type']=oe.fit_transform(data['type'].values.reshape(-1,1))
#we can perform the above step if we more than one type of avacado's in our data


data['Date']=oe.fit_transform(data['Date'].values.reshape(-1,1))
data['region']=oe.fit_transform(data['region'].values.reshape(-1,1))
data['year']=oe.fit_transform(data['year'].values.reshape(-1,1))
```

# Skewness and outliers removal:

```python
data.skew()
```

```python
#removing outliers
from scipy.stats import zscore
z=np.abs(zscore(data))
z
```

```python
#threshold=3
print(np.where(z>3))
```

```python
data_new=data[(z<3).all(axis=1)] #removing outliers
```

```python
data.shape
```

```python
data_new.shape
```

```python
#Percentage data loss:
loss_percent=(1517-1436)/1517*100
print(loss_percent)
```

```python
x=data_new.drop('AveragePrice', axis=1)
y=data_new['AveragePrice']
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import power_transform
data_new=power_transform(x)

data_new=pd.DataFrame(data_new,columns=x.columns)
```

```python
data_new.skew()
```

We can see, now skewness has been reduced.

# Let's perform scaling now:

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(data_new)
```

```python
x=data_new
```

```python
maxAcc=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    Ln=LinearRegression()
    Ln.fit(x_train,y_train)
    pred=Ln.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
print("Best accuracy is ",maxAcc, " at Random State ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=maxRS)
```

```python
model=[LinearRegression(),RandomForestRegressor(),AdaBoostRegressor(),Lasso(alpha=0.0001),Ridge(alpha=0.0001),ElasticNet(alpha=0.0001)]
for m in model:
    m.fit(x_train,y_train)
    #sc=m.score(x_train,y_train)
    predm=m.predict(x_test)
    acc=r2_score(y_test,predm)
    print('Accuracy Score of',m,'is:',acc)
    print('mean_absolute_error:',mean_absolute_error(y_test,predm))
    print('mean_squared_error:',mean_squared_error(y_test,predm))
    print('Root mean_squared_error:',np.sqrt(mean_squared_error(y_test,predm)))
    print("\n")
```

# Cross Validation of model:

```python
#from sklearn.metrics import make_scorer
#scoring={'r2':'r2_score'}
model=[LinearRegression(),RandomForestRegressor(),AdaBoostRegressor(),Lasso(alpha=0.0001),Ridge(alpha=0.0001),ElasticNet(alpha=0.0001)]
for m in model:
    score=cross_val_score(m,x,y,cv=5)
    print("Score for",m,"is: ",score.mean())
```
